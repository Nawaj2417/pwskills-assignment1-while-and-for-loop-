```python
Q2. State and Explain the features of MongoDB.
Ans:- The features of MongoDb are :-
    1. Document-oriented :- MongoDB is a document-oriented database system 
        that stores data in flexible, JSON-like documents which can have varying structures.
    2. Schema-less: MongoDB is schema-less, meaning that it does not enforce a strict schema on the data. This makes it easy to add new fields and modify existing ones without having to change the schema.
    3. Indexing: MongoDB supports indexing of any field in a document, which makes querying faster and more efficient.
    4. Aggregation: MongoDB offers a powerful aggregation framework that allows for complex data processing, such as grouping, filtering, and joining data
    5. High availability: MongoDB offers built-in replication and sharding, which provides high availability and scalability for large-scale applications.
```

```python
Q4. Using the database and the collection created in question number 3, write a code to insert one record,
and insert many records. Use the find() and find_one() methods to print the inserted record.
Ans:- 
    
import pymongo

# establish a connection to the MongoDB server
client = pymongo.MongoClient("mongodb+srv://Digicode:digicode@cluster0.rwamhl5.mongodb.net/?retryWrites=true&w=majority")
# create a new database called 'Digicode'
db = client['Digicode']
# create a new collection called 'coll_digicode' in the 'mydatabase'
coll_digicode =db["my_record"]


# insert one record into the collection
myrecord = {"name": "John", "age": 25, "city": "New York"}
coll_digicode.insert_one(myrecord)

# insert many records into the collection
myrecords = [
    {"name": "Alice", "age": 30, "city": "London"},
    {"name": "Bob", "age": 35, "city": "Paris"},
    {"name": "Charlie", "age": 40, "city": "Tokyo"}
]
coll_digicode.insert_many(myrecords)

# find the inserted record using find_one() method
result = coll_digicode.find_one({"name": "John"})
print(result)

# find all the inserted records using find() method
results = coll_digicode.find({})
for i in results:
    print(i)
```

```python
Q6.Explain why delete_one(), delete_many(), and drop() is used.
Ans:- delete_one(), delete_many() and drop() are used to remove data from a collection of database.
    
    'delete_one()' is used to remove a single document from a collection that matches a specified
    filter.
    ex:-
        db.users.delete_one({name: 'nawaj'})
        
    'delete_many' is used to remove multiple documents from a collection that matches 
    a specified filter.
    ex:-
        db.users.delete_many({age: {$gt: 30}})
    'drop()' is used to remove an entire collection or database.
    ex:-
        db.users.drop()


    
```
