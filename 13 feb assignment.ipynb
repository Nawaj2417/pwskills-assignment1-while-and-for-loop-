```python
# # Q1. Explain why we have to use the Exception class while creating a Custom Exception.
# # Note: Here Exception class refers to the base class for all the exceptions.
# Ans:- In Python, the Exception class is the base class for all the built-in exceptions, 
#     and it is recommended to use this class as the base class for creating custom exceptions.
    
#     Here is the some reasons to create a custom exception.
#     1. Inheritance:- Using the Exception class as the base class allows the custom exception to inherit all the properties and methods of the Exception class.
#     2. Catching :- . By using the Exception class as the base class, the custom exception will be caught by the same except statements that catch all other built-in exceptions.
#     3  Clarity: By creating custom exceptions that are based on the Exception class, you make your code more clear and readable.
```

```python
# Q2. Write a python program to print Python Exception Hierarchy.

def hierachy(exception, a=0):
    print(' ' * a + exception.__name__)
    for i in exception.__subclasses__():
        hierachy(i, a+1)
hierachy(Exception)
```

```python
Q5. Explain importError. What is ModuleNotFoundError ?
Ans:- "ImportError" is a built-in exception class in python that is raised when an import statements
    fails to find and load a module.
    
    For examples, suppose you have a module called "Error" that is located in a file called "error.py" . you can
    import this module by using "import " as "import error" . if this module is not found in the expected file
    or location then it raised "ImportError".
    
    similarly, "ModuleNotFoundError"  is raised when python cannot find the module you are trying to import.
    This error typically occurs when you misspell the name of the module or when the module is not installed or
    not present in the current directory.
```

```python
Q6. List down some best practices for exception handling in python.
Ans:- Some best practices for exception handling in python are:- 
    1. Instead of using a broad exceptions like "Exception", use specific exception types.
    2. Use "try-except" blocks to catch exceptions in your code.
    3. Use "finally" blocks to ensure that resources are cleaned up even if an exception is raised.
    4. Use a logging library to log exceptions and other errors in a code.
    5.Avoid 'try-except' blocks that catches all exceptions.
    
```
